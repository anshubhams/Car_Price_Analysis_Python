```python
# IMPORTING THE REQUIRED LIBS
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
#
```

```python
df=pd.read_csv('data.csv')
df.head(5)
```

```python
# 1. Rename columns for easier access (e.g., lowercase and underscores)
df.columns = df.columns.str.lower().str.replace(' ', '_')
df.head(5)
```

```python
df.info()
```

```python
# 2. Drop exact duplicate rows
df = df.drop_duplicates()
df.info()
```

```python
# 3. Find missing values
print(df.isnull().sum())
```

```python
#4. Drop irrelevant columns to simplify data model
df = df.drop(['market_category', 'engine_fuel_type'], axis=1)
```

```python
print(df.isnull().sum())
df.info()
df.describe()
```

```python
# 5. Drop empty rows (rows that have empty cells)
df = df.dropna()
df.info()
```

```python
after = len(df)
print(after)
```

```python
#6. Check the new size of your data
print(df.shape)
```

```python
#7. Creating a box plot via SNS to plot the price, to visualise the price distribution
plt.figure(figsize=(9, 4))
sns.boxplot(x=df['msrp'])
plt.title('Identifying Price Outliers')
plt.xlabel('Price')
plt.show()
```

```python
#8. Scatter Plot showing the relationship between Horsepower and Price
plt.figure(figsize=(8, 4))
sns.scatterplot(x=df['engine_hp'], y=df['msrp'])
plt.title('Relationship between Horsepower and Price')
plt.xlabel('Horsepower')
plt.ylabel('Price')
plt.show()
```

```python
#9. Bar graph (countplot) for Makers v Number of cars.
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(10, 6))
sns.countplot(x='make', data=df, order=df['make'].value_counts().index)
plt.title('Number of Cars by Make')
plt.xlabel('Car Make')
plt.ylabel('Number of Cars')
plt.xticks(rotation=90)
plt.show()
```

```python
#10. Correlation heatmap of car features
numerical_df = df.select_dtypes(include=['number'])
corr = numerical_df.corr()

plt.figure(figsize=(8, 6  ))
sns.heatmap(corr, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Heatmap of Car Features')
plt.show()
```

```python
#11. Categorical Analysis
## Count how many times each brand (Make) appears in the data:
brand_counts = df['make'].value_counts()

## Select the top 10 brands:
top_10_brands = brand_counts.head(10)

## Create a bar chart to visualise it:
plt.figure(figsize=(8, 5))
top_10_brands.plot(kind='bar')
plt.title('Top 10 Car Brands in this Dataset')
plt.xlabel('Brand')
plt.ylabel('Number of Cars')
plt.show()
```

```python
#12. Grouped Analysis - Price v Transmission
# Find the avg price for each type of transmission
transmission_avg = df.groupby('transmission_type')['msrp'].mean()
print(transmission_avg.round(2))
```
